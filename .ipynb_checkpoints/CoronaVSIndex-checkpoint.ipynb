```python
GDPdf = pd.read_csv("GDP_0.csv")[["Country Name","2018"]].dropna().sort_values(by="2018", ascending=False)
World = GDPdf.loc[257]["2018"]
GDPdf = GDPdf.reset_index()[13::]
GDPdf = GDPdf.drop(range(14,18))
GDPdf = GDPdf.drop(range(19,24))
GDPdf = GDPdf.drop(25)
GDPdf = GDPdf.drop(range(27,31))
GDPdf = GDPdf.drop([33,35,39,40,41,43,47,48,50,52,54,57,59,63,67])
```

```python
def getGdf(mainDF, totalGDP, gtopNr):
    tempGdf = mainDF.sort_values(by="2018", ascending=False).head(gtopNr)
    tempGdf = tempGdf[["Country Name","2018"]].reset_index()
    tempGdf = tempGdf.drop(columns=["index"])
    GTotal = tempGdf["2018"].sum()
    addDF = pd.DataFrame({"Country Name": ["Rest"], "2018":[totalGDP-GTotal] })
    tempGdf = tempGdf.append(addDF).reset_index()
    tempGdf = tempGdf.drop(columns=["index"])
    tempGdf["Ratio"] = tempGdf["2018"]/totalGDP
    return tempGdf
```

```python
G10df = getGdf(GDPdf, World, 50)
G10CountryList = G10df["Country Name"][0:50].replace("Korea, Rep.", "South Korea").replace("Russian Federation","Russia")
```

```python
def avCoronaDF(df, binNr):
    testdf = df
    dflen = len(testdf)
    totalCaseNEW = [testdf["total_cases"][binNr*i:binNr*(i+1)].mean() for i in range(int(dflen/binNr))]
    newCaseNEW = [testdf["new_cases"][binNr*i:binNr*(i+1)].mean() for i in range(int(dflen/binNr))]
    return [totalCaseNEW, newCaseNEW]
```

#### 1.3.1 Optimise Model with CHN and KOR

In China data, there is huge spike, due to change of method of estimation. Let us ignor that for now.

```python
CHNData0 = getData("China",100)
CHNData=CHNData0[CHNData0["new_cases"]<4000]
KORData = getData("South Korea",100)
```

Luckly, Python has skewed Gaussian predefined.

```python
from scipy.stats import skewnorm
def skFct(t,t0,eta, Amp,a):
    return Amp*skewnorm.pdf(abs(eta)*(t-t0), abs(a))

def getSkParam(ctryData,fitfct,p0list):
    xdata = np.arange(len(ctryData))
    ydata = ctryData["new_cases"].tolist()
    popt, pcov = curve_fit(fitfct, xdata, ydata,p0=p0list,maxfev=10000)
    return [xdata,ydata,popt]

def getSkFitRes(getSkParamOut,a):
    temp = getSkParamOut[1]-skFct(getSkParamOut[0], *getSkParamOut[2][0:3],a)
    return np.std(temp)

def getSkFitRes0(getSkParamOut):
    temp = getSkParamOut[1]-skFct(getSkParamOut[0], *getSkParamOut[2])
    return np.std(temp)

def plotSkFit(cntryRes, fitfct, linelabel):
    xdata = np.arange(len(cntryRes[0])+20)
    ydata = cntryRes[1]
    param = cntryRes[2]
    yfitdata = fitfct(xdata, *param)
    tempdf = pd.DataFrame({"days":np.arange(len(cntryRes[0])),"ydata": ydata})
    tempfitdf = pd.DataFrame({"days":xdata, "fitdata": yfitdata})
    tempdf.plot(x="days",y="ydata",ax=ax,label=linelabel)
    tempfitdf.plot(x="days",y="fitdata",ax=ax, legend=False, style = "--")
    plt.xlabel("days")
    plt.ylabel("New Cases")
```

```python
CHNRes = getSkParam(CHNData,skFct,[10,0.1,4000,2])
KORRes = getSkParam(KORData,skFct,[10,0.1,1000,2])
fig, ax = plt.subplots(figsize=(8,4))
plotSkFit(CHNRes,skFct,"China")
plotSkFit(KORRes,skFct, "Korea")
```

I use the mean for skewness a=2.3. <br>
We use eta and amp as fit parameter.<br>
AMP is related to the density of the contry.<br>
eta is realted how striclt goverment's measure.

```python
a0=np.mean([CHNRes[2][-1],KORRes[2][-1]])
```

```python
def showCountry(ctr, caseTH,maxCase, maxDeath,indexDF, indexLabel):
    coviddata = getData(ctr,caseTH)
    fig, ax = plt.subplots(nrows=3,ncols=2, figsize=(16,12))
    coviddata.plot(x="date",y="new_cases", logy=False, label=ctr, ax=ax[0,0])
    (indexDF["2020"::]*maxCase/max(indexDF["Mean"])).plot(y="Mean", ax=ax[0,0])
    ax[0,0].legend(["COVID19-NewCases",indexLabel])

    ax[0,1].scatter(x=coviddata["new_cases"],y=indexDF.iloc[-len(coviddata)::]["Mean"])
    ax[0,1].set_xlabel("NewCases")
    ax[0,1].set_ylabel(indexLabel)

    coviddata.plot(x="date",y="new_deaths", logy=False, label=ctr, ax=ax[1,0])
    (indexDF["2020"::]*maxDeath/max(indexDF["Mean"])).plot(y="Mean", ax=ax[1,0])
    ax[1,0].legend(["new_deaths",indexLabel])

    ax[1,1].scatter(x=coviddata["new_deaths"],y=indexDF.iloc[-len(coviddata)::]["Mean"])
    ax[1,1].set_xlabel("NewDeaths")
    ax[1,1].set_ylabel(indexLabel)

    indexDF["2020"::].plot(y="Vola", ax=ax[2,1])
    indexDF["2020"::].plot(y="Mom", ax=ax[2,1])
    ax[2,1].legend(["Volatity","Momentum"])
    return plt.show()
```
